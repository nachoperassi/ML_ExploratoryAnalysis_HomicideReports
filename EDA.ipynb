# Dataset: Homicide Reports, 1980-2014

The Murder Accountability Project is the most complete database of homicides in the United States currently available. This dataset includes murders from the FBI's Supplementary Homicide Report from 1976 to the present and Freedom of Information Act data on more than 22,000 homicides that were not reported to the Justice Department. This dataset **includes the age, race, sex, ethnicity of victims and perpetrators, in addition to the relationship between the victim and perpetrator and weapon used**.

The data was compiled and made available by the Murder Accountability Project, founded by Thomas Hargrove.

# Balance of the output variable based on the age of the victim

The **victim_age** column contains 101 unique values: (0 - 99) and 998.

Me made a web search and found that the 998 value represents an unknown victim, so we decided to ignore that value.

To make the problem easier and have a cleaner way to graph the data, we decided to create a new column called **victim_age_group** with **four unique values**:

** Child: (0-14) **

** Youth: (15-24) **

** Adult: (25-64) **

** Elderly: (65-99) **

And we set an 'Unknown' value in the cases where the victim_age is 998.

```python
ds2 = ds.copy()

victim_age_group = []

for row in ds2.victim_age:
    if row <= 14:
        victim_age_group.append('Child')
    elif row <= 24:
        victim_age_group.append('Youth')
    elif row <= 64:
        victim_age_group.append('Adult')
    elif row <= 99:
        victim_age_group.append('Elderly')
    else:
        victim_age_group.append('Unknown')

ds2['victim_age_group'] = victim_age_group

ds2.victim_age_group.value_counts()
```

```python
ncols = ds2[ds2.victim_age_group != 'Unknown'].victim_age_group.nunique()

fig, axes = plt.subplots(nrows=1, ncols=ncols, figsize=(16,4))

for c, i in zip(ds2[ds2.victim_age_group != 'Unknown'].victim_age_group.unique(), range(ncols)):
    ds2[ds2.victim_age_group == c].crime_solved.value_counts().plot.pie(autopct='%1.0f%%', ax=axes[i])
    axes[i].set_title(c)
```

**The cases in which the victim age group is 'Elderly' have a lower percentage of resolution and the cases in  which the victim age group is 'Child' have a notoriously higher percentage of resolution**
